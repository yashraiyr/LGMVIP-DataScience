```python
# Importing libraries
import io
import numpy as np
import pandas as pd
import seaborn as sb
from sklearn import tree
from google.colab import files
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse
```

```python
# Reading the dataset
data = pd.read_csv(io.BytesIO(uploaded['Iris.csv']))
# Printing the first 5 rows of the dataset
data.head()
```

```python
# Grouping Features
#Assigning features into one group
X=data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']] 
#Assigning Prediction outcome to one group
Y=data['Species'] 
```

```python
# Splitting the data into train(80%) and test(20%)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
```

```python
# Decion Tree Classifier
decisiontree = DecisionTreeClassifier(max_depth=6, random_state=1)
decisiontree.fit(X_train, Y_train) # Fit dt to the training set
y_pred = decisiontree.predict(X_test) # Predict test set labels
print(y_pred[0:25])
```

```python
# Visualizing Descision Tree Classifier
plt.figure(figsize=(150,150))
tree.plot_tree(decisiontree,fontsize=15)
```

```python
# Evaluate Model
train_score_tree = str(decisiontree.score(X_train, Y_train) * 100)
test_score_tree = str(decisiontree.score(X_test, Y_test) * 100)
print(f'Train Score : {train_score_tree[:5]}%\nTest Score : {test_score_tree[:5]}%')
```
